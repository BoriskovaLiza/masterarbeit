### Phi functions validation and comparison

```python
import mySDC
import numpy as np
from matplotlib import pyplot as plt

print(mySDC.get_phi_by_contour_integration(0., 1))
```

```python
X = np.linspace(-1., 1., 1000)

for n in range(6):
    expl = np.array([mySDC.get_phi_by_explicit_formula(x, n) for x in X])
    series = np.array([mySDC.get_phi_by_series(x, n) for x in X])
    contour = np.array([mySDC.get_phi_by_contour_integration(x, n) for x in X])
    
    mySDC.plot_phi('Phi no %d' % n, ["explicit", "series", "integral"], \
                      [expl, series, contour], X)
```

```python
X = np.linspace(-5.j, 5.j, 1000)

for n in range(6):
    expl = np.array([mySDC.get_phi_by_explicit_formula(x, n) for x in X], dtype="complex128")
    series = np.array([mySDC.get_phi_by_series(x, n) for x in X], dtype="complex128")
    contour = np.array([mySDC.get_phi_by_contour_integration(x, n) for x in X], dtype="complex128")
    
    mySDC.plot_phi('Phi no %d' % n, ["explicit", "series", "integral"], \
                      [expl, series, contour], X * -1.j)
```

```python
X = np.linspace(-.5, .5, 1000)

for n in range(6):
    diff = np.array([mySDC.get_phi_by_contour_integration(x, n) - mySDC.get_phi_by_series(x, n) for x in X], \
                    dtype="complex128")
    mySDC.plot_phi('Phi no %d' % n, ["diff(series, integral)"], [diff], X)
```

```python
X = np.linspace(-.5j, .5j, 1000)

for n in range(6):
    diff = np.array([mySDC.get_phi_by_contour_integration(x, n) - mySDC.get_phi_by_series(x, n) for x in X], \
                    dtype="complex128")
    mySDC.plot_phi('Phi no %d' % n, ["diff(series, integral)"], [diff], X * -1.j)
```
