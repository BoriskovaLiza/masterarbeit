### Phi functions validation and comparison

```python
import mySDC
import numpy as np
from matplotlib import pyplot as plt

print(mySDC.get_phi_by_contour_integration([0.], 1))
```

```python
for n in range(6):
    mySDC.plot_functions('Phi no %d' % n, ["explicit", "series", "integral"], \
                      [lambda z: mySDC.get_phi_by_explicit_formula(z, n), \
                        lambda z: mySDC.get_phi_by_series(z, n), \
                        lambda z: mySDC.get_phi_by_contour_integration(z, n)], np.linspace(-1., 1., 1000))
```

```python
for n in range(6):
    mySDC.plot_functions_imag('Phi no %d' % n, ["explicit", "series", "integral"], \
                      [lambda z: mySDC.get_phi_by_explicit_formula(z, n), \
                        lambda z: mySDC.get_phi_by_series(z, n), \
                        lambda z: mySDC.get_phi_by_contour_integration(z, n)], np.linspace(-5.j, 5.j, 1000))
```

```python
for n in range(6):
    mySDC.plot_functions('Phi no %d' % n, ["diff(series, integral)"], \
                      [lambda z: mySDC.get_phi_by_contour_integration(z, n) - mySDC.get_phi_by_series(z, n)], np.linspace(-.5, .5, 1000))
```

```python
for n in range(6):
    mySDC.plot_functions_imag('Phi no %d' % n, ["diff(series, integral)"], \
                      [lambda z: mySDC.get_phi_by_contour_integration(z, n) - mySDC.get_phi_by_series(z, n)], np.linspace(-.5j, .5j, 1000))
```
