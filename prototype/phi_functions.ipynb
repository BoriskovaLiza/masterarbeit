### Phi functions validation and comparison

```python
import mySDC
import numpy as np
from matplotlib import pyplot as plt

for n in range(6):
    mySDC.plot_functions('Phi no %d' % n, ["explicit", "series", "integral"], \
                      [lambda z: mySDC.get_phi_by_explicit_formula(z, n), \
                        lambda z: mySDC.get_phi_by_series(z, n), \
                        lambda z: mySDC.get_phi_by_contour_integration(z, n)], np.linspace(-0.05, 0.05, 1000))
```

```python
for n in range(6):
    mySDC.plot_functions_imag('Phi no %d' % n, ["explicit", "series", "integral"], \
                      [lambda z: mySDC.get_phi_by_explicit_formula(z, n), \
                        lambda z: mySDC.get_phi_by_series(z, n), \
                        lambda z: mySDC.get_phi_by_contour_integration(z, n)], np.linspace(-0.05j, 0.05j, 1000))
```

```python
for n in range(6):
    mySDC.plot_functions('Phi no %d' % n, ["diff(series, integral)"], \
                      [lambda z: mySDC.get_phi_by_contour_integration(z, n) - mySDC.get_phi_by_series(z, n)], np.linspace(-0.05, 0.05, 1000))
```

```python
for n in range(6):
    mySDC.plot_functions_imag('Phi no %d' % n, ["diff(series, integral)"], \
                      [lambda z: mySDC.get_phi_by_contour_integration(z, n) - mySDC.get_phi_by_series(z, n)], np.linspace(-0.05j, 0.05j, 1000))
```

```python
print("Relative error @ 1e-10j:")
for n in range(8):
    print('Phi no %d' % n)
    print(mySDC.get_phi_by_contour_integration(1e-10j, n), mySDC.get_phi_by_series(1e-10j, n), (mySDC.get_phi_by_contour_integration(1e-10j, n) - mySDC.get_phi_by_series(1e-10j, n))/mySDC.get_phi_by_series(1e-10j, n))
```
